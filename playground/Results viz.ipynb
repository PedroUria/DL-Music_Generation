```python
import matplotlib.pyplot as plt
import os
```

```python
def plot_losses(filepath):
    f = open(filepath,'r')
    results = f.readlines()
    epoch = []
    loss = []
    for line in results:
        if 'epoch number' in line:
            epoch.append(int(line.split(':')[1].split('-')[0]))
        if 'loss: ' in line:
            loss.append(float(line.split(':')[1][1:-2]))
    plt.figure(figsize=(15,10))
    plt.plot(epoch, loss)
    plt.xlabel('Epochs', fontsize=13)
    plt.ylabel('Loss', fontsize=13)
    plt.title('Loss v Epochs ['+filepath.split('.')[0].split('_')[0]+']', fontsize=18)
    plt.savefig('result_'+filepath.split('.')[0].split('_')[0]+'.png')
```

```python
plot_losses('haydn_results.txt')
```

```python
plot_losses('mendelssohn_results.txt')
```

```python
plot_losses('schubert_results.txt')
```
